```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.pandas.set_option('display.max_columns', None)
```

## Importing Train and Test data

```python
train_df_upd = pd.read_csv('train_df_upd.csv')
test_df_upd = pd.read_csv('test_df_upd.csv')
```

```python
X_train = train_df_upd.iloc[:,:-1].values
y_train = train_df_upd.iloc[:,-1].values
X_test = test_df_upd.iloc[:,:-1].values
y_test = test_df_upd.iloc[:,-1].values
```

## Training Regressor

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
tuned_parameters = [{'kernel': ['linear'], 'C': [400,500,600,700,800]},
                   {'C': range(100,1000,100), 'gamma': np.linspace(0,1,11), 'kernel': ['rbf']}]
scores = ['precision', 'recall']
grid_search = GridSearchCV(estimator = SVR(),
                           param_grid = tuned_parameters,
                           cv = 15,
                           n_jobs = -1)
grid_search = grid_search.fit(X_train, y_train)
best_score = grid_search.best_score_
best_parameters = grid_search.best_params_
print("Best Score: {:.2f} %".format(best_score*100))
print("Best Parameters:", best_parameters)
```

```python
y_pred = grid_search.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_pred)
print(r2_score)
```

```python
regressor = SVR(C=500, kernel='linear',random_state=32)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
